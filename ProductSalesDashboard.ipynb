```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df=pd.read_csv("product.csv")
print(df)
df.head()
```

1. Total Sales by Category.. 

```python
total_sales=df.groupby("Category")["Sales"].sum().reset_index()
print(total_sales)
```

```python

sns.barplot(data=total_sales, x="Category", y="Sales",palette="Set1")

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

2. Sales Distribution..

```python
sns.histplot(df["Sales"])
plt.show()
```

3. kis year mein kitna profit hua hai

```python
df["Date"]=pd.to_datetime(df["Date"])
df.groupby(df["Date"].dt.year)["Sales"].sum().reset_index()

```

```python
df["Date"] = pd.to_datetime(df["Date"])

year_sales = df.groupby(df["Date"].dt.year)["Sales"].sum().reset_index()

sns.lineplot(data=year_sales, x="Date", y="Sales", marker="o")
plt.grid(True)
plt.title("Year-wise Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

4.  Create a bar plot for average sales

```python
avg_sales=df.groupby("Category")["Sales"].mean().reset_index()
print(avg_sales)
```

```python
sns.barplot(data=avg_sales,x="Category",y="Sales",palette="Set2")
plt.title("Average Sales",fontweight="bold")
plt.xlabel("Category",fontweight="bold",color="Purple")
plt.ylabel("Sales",fontweight="bold",color="Purple")

plt.show()
```

5. create a  plot  for categories profit 

```python
total_profit=df.groupby("Category")["Profit"].sum().reset_index()
print(total_profit)
```

```python
sns.lineplot(data=total_profit,x="Category",y="Profit",marker="o")
plt.grid(True)
plt.show()
```

Region wise total sales

```python
region_sales = df.groupby("Region")["Sales"].sum().reset_index()

print(region_sales)
```

```python
plt.pie(region_sales.values,labels=region_sales.index,autopct="%1.1f%%", 
        colors=sns.color_palette("dark"))

plt.title("Sales Distribution by Region",fontweight="bold",color="Purple")
plt.show()
```
