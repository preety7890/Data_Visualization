```python
plt.figure(figsize=(8,5))
sns.countplot(x="Department",data=df,palette="Set2")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Salary"],color="green")
plt.title("Salary distribution")
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.scatterplot(x="Salary",y="Experience",data=df,hue="Department")
plt.title("Salary vs Experience")
plt.show()
```

```python
total=df.groupby("Department")["Employee_ID"].size().reset_index(name="total_employee")
print(total)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=total,x="Department",y="total_employee",palette="Set1")
plt.title("Total employee in each department")
plt.show()
```

```python
avg_sal=df.groupby("Department")["Salary"].mean().reset_index(name="Avg_salary")
print(avg_sal)
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(avg_sal,marker=">")
plt.grid(True)
plt.show()
```
