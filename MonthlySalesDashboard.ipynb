```python
import pandas as pd
import matplotlib.pyplot as plt

months=['Jan','Feb','March','Apr','May','Jun','July','Aug','Sept','Oct','Nov','Dec']
sales=[50000,62000,70000,65000,71000,56000,73000,69000,78000,75000,77000,80000]

# line chart

plt.plot(months,sales,color='blue',linestyle='--',marker='o')
plt.grid(True)
plt.xlabel("Months",color='red')
plt.ylabel("sales",color='red')
plt.savefig("sales_report.png")
plt.show()

```

```python
plt.bar(months,sales,color='blue')
plt.xlabel("Months",color='red')
plt.ylabel("sales",color='red')
plt.savefig("sales_report1.png")
plt.show()
```

```python
# pie chart
plt.figure(figsize=(7,7))

plt.pie(sales,labels=months,autopct='%1.1f%%')

plt.title("Monthly Sales Distribution")
plt.show()
```
