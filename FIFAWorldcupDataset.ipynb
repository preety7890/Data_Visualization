```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/home/Downloads/matches.csv")
print(df)
df.head()
```

1. Total matches played in each Round ? 

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x='round', palette='viridis')

plt.title("Total Matches Played in Each Round", fontweight="bold",color="Purple")
plt.xlabel("Round",color="red",fontweight="bold")
plt.ylabel("Number of Matches",color="red",fontweight="bold")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

2. Matches played on each day of the week

```python
week=df.groupby("dayofweek").size().reset_index(name="Matches")
print(week)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=week,
    x="dayofweek",
    y="Matches",
    palette="Set1"
)

plt.title("Matches Played on Each Day of the Week",fontweight="bold",color="Blue")
plt.xlabel("Day of Week",fontweight="bold",color="red")
plt.ylabel("Number of Matches",fontweight="bold",color="red")

plt.show()
```

3. Distribution of attendance

```python
attendance=df["attendance"]
print(attendance)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(attendance,color="darkcyan")
plt.title("Distribution of attendance",fontweight="bold",color="Green")
plt.tight_layout()
plt.xticks(rotation=80)
plt.show()
```

4. Which stadium hosted the most matches?

```python
most_matches=df.groupby("venue").size().reset_index(name="matches")
print(most_matches)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=most_matches,x="venue",y="matches",palette="Set1")
plt.title("Stadium most matches",fontweight="bold",color="Purple")
plt.xlabel("Venue",fontweight="bold",color="green")
plt.ylabel("no of matches",fontweight="bold",color="green")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

5. Top 10 referees who conducted the most matches.

```python
top=df.groupby("referee").size().sort_values(ascending=False).head(10).reset_index(name="matches")
print(top)
```

```python
sns.lineplot(data=top,x="referee",y="matches",color="red",marker=">")
plt.title("Top 10 Referees ",fontweight="bold",color="Green")
plt.grid(True)
plt.xlabel("Referee",fontweight="bold",color="purple")
plt.ylabel("Matches",fontweight="bold",color="purple")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

6. Home possession vs Away possession.

```python
sns.scatterplot(data=df,x="home_possession",y="away_possession",hue="round",s=80)
plt.title("Home Possession vs Away Possession",fontweight="bold",color="Purple")
plt.tight_layout()
plt.show()
```

7. Home yellow cards distribution vs away yellow cards distribution

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x="home_cards_yellow",y="away_cards_yellow",hue="round")
plt.title("Home Yellow cards distribution VS Away Yellow cards distribution",color="purple")
plt.show()
```

8. Top 10 teams with highest shots on target.

```python

away = df.groupby("away_team")["away_sot"].sum().sort_values(ascending=False).head(10).reset_index(name="sot")
print(away)
```

```python
home = df.groupby("home_team")["home_sot"].sum().sort_values(ascending=False).head(10).reset_index(name="sot")
print(home)
```

```python
plt.figure(figsize=(14,5))

# Away
plt.subplot(1,2,1)
plt.bar(away["away_team"], away["sot"],color="blue")
plt.title("Top 10 Away Teams (Shots on Target)",fontweight="bold")
plt.xticks(rotation=45)
plt.xlabel("Team")
plt.ylabel("SOT")

# Home
plt.subplot(1,2,2)
plt.bar(home["home_team"], home["sot"], color="green")
plt.title("Top 10 Home Teams (Shots on Target)",fontweight="bold")
plt.xticks(rotation=45)
plt.xlabel("Team")
plt.ylabel("SOT")

plt.tight_layout()
plt.show()
```

9. Top 10 home team score...(goals)

```python
score=df.groupby("home_team")["home_score"].sum().sort_values(ascending=False).head(10).reset_index(name="score")
print(score)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=score ,x="home_team",y="score",palette="viridis")
plt.title("Top 10 Home Team score (Goals)",fontweight="bold",color="blue")

plt.tight_layout()
plt.show()
```

10. Top 10 away team score(goals)

```python
goal=df.groupby("away_team")["away_score"].sum().sort_values(ascending=False).head(10).reset_index(name="goals")
print(goal)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(data=goal,x="away_team",y="goals",palette="Set1")
plt.title("Top 10 Away team score (Goals)",fontweight="bold",color="Green")
plt.tight_layout()
plt.show()
```
