```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("employeedata.csv")
print(df)
```

```python
plt.figure(figsize=(7,4))
sns.countplot(
    data=df,
    x="Department"
 
)
plt.title("Employee Count by Department")
plt.show()
```

# 2. Average Salary by Department


```python
salary_avg = df.groupby("Department")["Salary"].mean().reset_index()
print(salary_avg)
```

```python
plt.figure(figsize=(7,4))

sns.lineplot(
    data=salary_avg,
    x="Department",
    y="Salary",
    marker="o",
    color="red",
    linestyle="--"
)
plt.grid(True)
plt.title("Average Salary by Department")
plt.ylabel("Average Salary")
plt.show()
```

```python
# 3. Salary Distribution

plt.figure(figsize=(7,4))

sns.histplot(
    df["Salary"],
    bins=5,
    color="blue"
)

plt.title("Salary Distribution")
plt.show()
```

 4. Scatter Plot - Experience vs Salary.

```python

plt.figure(figsize=(7,4))
sns.scatterplot(
    data=df,
    x="Experience",
    y="Salary",
    hue="Department"
)

plt.title("Experience vs Salary",fontweight="bold",color="Purple")
plt.show()
```

```python
# 5. Box Plot - Salary Distribution by Department
plt.figure(figsize=(7,4))
sns.boxplot(
    data=df,
    x="Department",
    y="Salary",
    palette="Set1"
)

plt.title("Salary Spread by Department")
plt.show()
```
